### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
cars = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]
most_common_origin = cars['origin'].mode().values[0]
print(most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here
cars = df[df['origin'] == 'japan']
mean_acceleration = round(cars['acceleration'].mean(), 2)
print(mean_acceleration)
```

### 5. Which year had the highest average mpg?

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
median_mpg = df['mpg'].median()
above_median_mpg_cars = df[df['mpg'] > median_mpg]
best_ratio = above_median_mpg_cars['horsepower_per_weight'].max()
best_cars = above_median_mpg_cars[above_median_mpg_cars['horsepower_per_weight'] == best_ratio]
for index, row in best_cars.iterrows():
    print(f"{row['name']} : {best_ratio:.4f}")
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
df = df.groupby(['origin', 'model_year'])['mpg'].mean().reset_index()
pivot_df = df.pivot(index='model_year', columns='origin', values='mpg')
plt.figure(figsize=(10,6))
for column in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[column], label=column)

plt.title('Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
df = pd.read_csv('/home/aravindmanoharan/DS_2025/Week 1 Assignment/Cars.csv')
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df,
                x=df['horsepower'],
                y=df['weight'],
                hue=df['origin'],
                size=df['mpg'],
                palette=['blue', 'green', 'red'],
                hue_order=['japan', 'europe', 'usa'])

# Set plot title and axis titles
plt.title('Relationship between Horsepower, Weight, MPG, and Origin')
plt.xlabel('Horsepower')
plt.ylabel('Weight')

# Show the legend with a custom label for size
plt.legend(title='Origin (Color) / MPG (Size)', bbox_to_anchor=(1.05, 1), loc=2)

# Show the plot
plt.tight_layout()
plt.show()
```
